### ðŸ”¹ What is a Scatter Plot?
A scatter plot is a type of data visualization used to show the relationship between two numerical variables.

Each point in the plot represents one observation (row) in the dataset.
- X-axis â†’ First numeric variable
- Y-axis â†’ Second numeric variable
- Each dot â†’ One data record

### ðŸ”¹ Why Do We Use a Scatter Plot?
1. A scatter plot helps us to:
2. Understand the relationship between two variables
3. Identify trends (positive or negative correlation)
4. Detect outliers
5. Observe clusters or patterns
6. Analyze distribution of data points

### ðŸ”¹ Scatter Plot Method in Seaborn
- `sns.scatterplot()`

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sk = sns.load_dataset("penguins")
sk
```

```python
sk.columns
```

```python
s ={"Male":"*", "Female":"o"}
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=sk, hue="sex", style="sex", size="sex", sizes=(100, 50),
                palette="gist_ncar", alpha=1, markers=s
            )
plt.show()
```
