# ðŸ”¹ What is a Line Plot in Seaborn?
A line plot in Seaborn is used to show the relationship between two continuous variables, usually to observe trends over time or changes in data.

### ðŸ”¹ Seaborn Line Plot Method
Seaborn, a line plot is created using the `lineplot()` function.

### ðŸ‘‰ Function Name
`sns.lineplot()`

### ðŸ”¹ Basic Syntax
`sns.lineplot(x, y, data=None)`

##### ðŸ”¹ Important Parameters 
- x â†’ Values for x-axis
- y â†’ Values for y-axis
- data â†’ Dataset (Pandas DataFrame)
- hue â†’ Draw multiple lines using categories
- style â†’ Different line styles
- markers â†’ Show markers on data points
- ci â†’ Confidence interval

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 3, 4, 1, 5, 6, 7]

xy = pd.DataFrame({"x":x, "y":y})

# seaborn chart
sns.lineplot(x=x, y=y, data=xy)

plt.show()
```

```python
# matplotlib
plt.plot(x, y)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

sns.lineplot(x=x, y=y)
plt.show()
```

# ðŸ”¹ Line Plot with DataFrame Example

```python
data = {
    "Year": [2018, 2019, 2020, 2021, 2022],
    "Sales": [200, 250, 300, 350, 400]
}

df = pd.DataFrame(data)

sns.lineplot(x="Year", y="Sales", data=df)
plt.show()
```

```python
sk = sns.load_dataset("penguins").head(20)
sk
```

```python
sns.lineplot(x="bill_length_mm", y="bill_depth_mm", data=sk, hue="sex", style="sex", palette="Accent", markers=["o", ">"], dashes=False, legend=False)

plt.grid()
plt.title("python")
plt.show()
```
