# ðŸ“Š Seaborn Histogram Plot Method in Python

#### ðŸ”¹ What is a Histogram in Seaborn?
A Histogram is used to show the distribution of numerical (continuous) data.
- How data is distributed
- Frequency of values
- Range where data values lie

### ðŸ”¹ Seaborn Histogram Function
histogram is created using the histplot() function.

#### ðŸ‘‰ Function Name
`sns.histplot()`

#### Basic Syntax
`sns.histplot(data, bins=None, kde=False)`

#### OR (with DataFrame)
`sns.histplot(x="column_name", data=df)`


### ðŸ”¹ Important Parameters (Theory)
- x â†’ Column or data values
- data â†’ Pandas DataFrame
- bins â†’ Number of bars (intervals)
- kde â†’ Kernel Density Estimation (smooth curve)
- color â†’ Histogram color
- palette â†’ Colors (when hue is used)
- hue â†’ Category-wise histogram
- multiple â†’ `layer`, `stack`, `dodge`, `fill`

## ðŸ”¹ Simple Histogram Example

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset("penguins")
df
```

### ðŸ”¹ What is displot()?
displot() is a figure-level function in Seaborn used to visualize the distribution of data.
- Histogram
- KDE plot
- ECDF plot
- Rug plot

### ðŸ”¹ Function Name
- `sns.displot()`

### ðŸ”¹ Basic Syntax
- `sns.displot(data)`

### With DataFrame:
- `sns.displot(data=df, x="column_name")`

### ðŸ”¹ Important Parameters
- data â†’ Dataset (Pandas DataFrame or array)
- x / y â†’ Column to plot
- kind â†’ Type of plot (hist, kde, ecdf)
- bins â†’ Number of bars (histogram)
- kde â†’ Add KDE curve (True / False)
- hue â†’ Category-wise distribution
- palette â†’ Colors
- col / row â†’ Multiple subplots

```python
sns.displot(df["flipper_length_mm"])
plt.show()
```

```python
sns.displot(df["flipper_length_mm"], bins=[170, 180, 190, 200, 210, 220, 230, 240])
plt.show()
```

```python
sns.displot(df["flipper_length_mm"], bins=[170, 180, 190, 200, 210, 220, 230, 240], kde=True)
plt.show()
```

```python
sns.displot(df["flipper_length_mm"], kde=True, rug=True, color="g", log_scale=True)
plt.show()
```
