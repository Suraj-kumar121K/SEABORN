### ðŸ”¹ What is a Heatmap?
A heatmap is a data visualization technique that shows values in a matrix format using colors.
- Higher values â†’ darker / stronger color
- Lower values â†’ lighter color
- Each cell represents a value

It is mainly used to visualize:
- Correlation matrix
- ivot tables
- Missing values
- Aggregated data

#### ðŸ”¹ Why Use a Heatmap?
- Identify patterns quickly
- Detect correlation between variables
- Compare multiple values at once

Understand intensity using colors

### ðŸ”¹ Seaborn Heatmap Function
- `sns.heatmap()`

# Code

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
s = np.linspace(1, 10, 20).reshape(4, 5)
s
```

```python
y={"fontsize":20, "color":"r"}
v = sns.heatmap(s, vmin=0, vmax=10, cmap="PuOr", annot=True, annot_kws=y, linewidth=10, linecolor="g", cbar=False, xticklabels=False)
v.set(xlabel="Python", ylabel="suraj")
sns.set(font_scale=5)
plt.show()
```
