# üìä Seaborn BarPlot Method

### 1Ô∏è‚É£ What is Bar Plot?
- Bar Plot categorical data ko visualize karta hai.
- Ye har category ka average (mean) value dikhata hai by default.

#### 2Ô∏è‚É£ Seaborn BarPlot Function
`sns.barplot(x=None, y=None, data=None, hue=None)`

#### Parameters:
- x -->  Categorical variable         
- y -->  Numerical variable         
- data  --> DataFrame                    
- hue --> Grouping variable (optional) 
- palette  --> Color style                  
- estimator --> Function like mean, sum etc. 


3Ô∏è‚É£ Basic Example

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample Data
data = {
    "Department": ["IT", "HR", "Sales", "IT", "HR", "Sales"],
    "Salary": [50000, 40000, 45000, 52000, 42000, 47000]
}

import pandas as pd
df = pd.DataFrame(data)

# Bar Plot
sns.barplot(x="Department", y="Salary", data=df)

plt.title("Average Salary by Department")
plt.show()
```

4Ô∏è‚É£ Hue Parameter (Grouping)

```python
sns.barplot(x="Department", y="Salary", hue="Department", data=df)
plt.show()
```

5Ô∏è‚É£ Change Colors (Palette)

```python
sns.barplot(x="Department", y="Salary", data=df, palette="Set2")
plt.show()
```

# 6Ô∏è‚É£ Change Aggregation
Agar sum chahiye:

```python
import numpy as np

sns.barplot(x="Department", y="Salary", data=df, estimator=np.sum)
plt.show()
```

```python
df = sns.load_dataset("penguins")
df
```

```python
sns.barplot(x="island", y="bill_length_mm", hue="sex", data=df)
plt.show()
```

- saturation=5.3 :- image light

```python
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=df, hue="sex", 
            order=order_1, hue_order=["Female", "Male"], ci=100, n_boot=3, palette="icefire", saturation=5.3)
plt.show()
```

```python
order_1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=df, hue="sex", 
            order=order_1, hue_order=["Female", "Male"], orient="v", saturation=100, errcolor="b", dodge=False, alpha=0.7)
plt.show()
```
