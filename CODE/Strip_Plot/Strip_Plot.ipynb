# üìä Seaborn Strip Plot Method 

## üîπ What is Strip Plot?
### Strip Plot ek categorical scatter plot hai jo:
- Individual data points show karta hai
- Categories ke andar distribution dikhata hai
- Overlapping avoid karne ke liye thoda ‚Äújitter‚Äù use karta hai

## üìå Syntax
`sns.stripplot(data=df, x="Category", y="Value")`

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
df
```

```python
sns.stripplot(x="day", y="total_bill", data=df)
plt.show()
```

# üéØ Important Parameters
- üîπ 1Ô∏è‚É£ jitter
#### Points overlap na kare iske liye use hota hai.

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="rocket_r", linewidth=1, edgecolor="r", jitter=10)
plt.show()
```

# üîπhue
Multiple category comparison ke liye used hota hai:

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex")
plt.show()
```

## üîπpalette
Color control karne ke liye:

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="rocket_r")
plt.show()
```

## üìä linewidth
line ki size ko mota ya patla karna

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="rocket_r", linewidth=2)
plt.show()
```

## üìå edgecolor
üëâ Shape ya marker ke border (kinare) ka color

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="rocket_r", linewidth=1, edgecolor="r")
plt.show()
```

# üîπsize
Point ka size change karne ke liye:

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="rocket_r", linewidth=1, edgecolor="r", size=10)
plt.show()
```

# üîπdodge=True
Hue ke case me side-by-side separation karta hai:

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="rocket_r", linewidth=1, edgecolor="r", size=10, dodge=True)
plt.show()
```

# üîπMarker
üëâ Marker ka matlab hai data point ka shape

```python
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="rocket_r", linewidth=1, edgecolor="r", size=10, marker="<")
plt.show()
```

# üîπalpha
üëâ Alpha transparency (opacity) control karta hai
#### Matlab:
- alpha = 1 ‚Üí fully visible (solid)
- alpha = 0 ‚Üí completely transparent (invisible)
- alpha = 0.5 ‚Üí half transparent
- Range hoti hai -> 0 se 1 ke beech

```python
m={"Male":"o", "Female":"<"}
sns.stripplot(x="day", y="total_bill", data=df, hue="sex", palette="rocket_r", linewidth=1, edgecolor="r", size=10, marker="<", alpha=0.5)
plt.show()
```
