# ðŸ“Š Seaborn Pair Plot Method in Python

### ðŸ”¹ What is a Pair Plot?
A Pair Plot is a Seaborn function that:
- Shows pairwise relationships between numerical variables
- Displays scatter plots for each variable combination
- Shows distribution plots (histogram or KDE) on the diagonal

### ðŸ”¹ Function Used
`sns.pairplot()`

### ðŸŽ¯ What Does Pair Plot Show?
If your dataset has 4 numeric columns:
- A
- B
- C
- D

###### Pairplot will automatically create:
- A vs B
- A vs C
- A vs D
- B vs C
- B vs D
- C vs D
##### On the diagonal:
- Distribution (histogram or KDE) of each variable
##### So it gives a complete visual relationship overview in one grid.

## ðŸ”¹ Basic Example

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset("tips")

sns.pairplot(df)
plt.show()
```

```python
df.columns
```

```python
sns.pairplot(df, hue="sex", hue_order=["Female", "Male"]) 
plt.show()
```

```python
sns.pairplot(df, vars=["total_bill", "tip"], hue="sex", hue_order=['Female', "Male"], palette='BuGn')
plt.show()
```

- `vars` = Only use these selected columns to create pairwise plots.

```python
sns.pairplot(df, hue="sex", hue_order=["Female", "Male"], kind="kde") # 'scatter', 'kde', 'hist', 'reg'
plt.show()
```

```python
sns.pairplot(df, hue="sex", hue_order=["Female", "Male"], kind="kde", diag_kind="hist") # 'scatter', 'kde', 'hist', 'reg'
plt.show()
```

```python
sns.pairplot(df, hue="sex", hue_order=["Female", "Male"], markers=["*", "<"]) # 'scatter', 'kde', 'hist', 'reg'
plt.show()
```
