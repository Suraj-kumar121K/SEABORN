# üéª Seaborn Violin Plot

### üîπ What is a Violin Plot?
A Violin Plot shows the distribution of numerical data across categories.

## üß† Theory Behind Violin Plot
### 1Ô∏è‚É£ Median
The white dot or central line.

### 2Ô∏è‚É£ Interquartile Range (IQR)
The thick bar inside

### 3Ô∏è‚É£ Density Shape
The outer curved shape shows how values are distributed.

### üîπ Method Name
`sns.violinplot()`

# üßæ Basic Syntax
`sns.violinplot(x=None, y=None, hue=None, data=None,
               palette=None, split=False, inner='box')`

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
df
```

```python
sns.violinplot(x="day", y="total_bill", data=df, inner="box")
plt.title("Total Bill Distribution by Day")
plt.show()
```

## üìä What This Shows
1. X-axis ‚Üí Day (Thur, Fri, Sat, Sun)
2. Y-axis ‚Üí Total bill amount

##### Each violin shows:
- Distribution shape
- Median
- Quartiles (inside box)

```python
sns.violinplot(x="day", y="total_bill",
               hue="sex",
               data=df,
               split=True,
               inner="box")

plt.title("Bill Distribution by Day and Gender")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=df, hue="time")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=df, linewidth=2, color="r")
plt.show()
```

```python
sns.violinplot(x="time", y='total_bill', data=df, order=["Dinner", "Lunch"], palette="Dark2", saturation=0.4)
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=df, hue="sex", split=True, scale="width")
plt.show()
```

```python
sns.violinplot(y="day", x="total_bill", data=df, hue="sex")
plt.show()
```
