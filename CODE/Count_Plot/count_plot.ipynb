# What's the difference between countplot and barplot?
- Countplot plots the count of the number of records by category.
- Barplot plots a value or metric for each category barplot plots the mean of a variable by category

# ðŸ“Š countplot() in Seaborn
### ðŸ”¹ What is countplot()?
- `countplot()` is a function in Seaborn that shows the count (frequency) of each category in a column.
- X-axis â†’ Categories
- Y-axis â†’ Count (number of occurrences)

#### ðŸ§¾ Basic Syntax

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sk = sns.load_dataset("tips")
sk
```

```python
sns.countplot(x="sex", data = sk, hue="smoker", color="r", saturation=(0.5))
plt.show()
```

```python
sns.barplot(x="sex", y="size", data=sk)
plt.show()
```
